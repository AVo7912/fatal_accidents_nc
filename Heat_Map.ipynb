```python
# import dependencies
import pandas  as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
import warnings
warnings.filterwarnings('ignore')
```

```python
# load csv data
df_acc = pd.read_csv(r"C:/Users/Aline/OneDrive/Documents/GitHub/fatal_accidents_nc/Resources/nc_accidents_data.CSV", encoding="ANSI")
```

```python
# create map object
map_hooray = folium.Map(location=[35.782, -80.793],
                    zoom_start = 7) 
```

```python
# make sure that data is in floats
df_acc['LATITUDE'] = df_acc['LATITUDE'].astype(float)
df_acc['LONGITUD'] = df_acc['LONGITUD'].astype(float)
```

```python
# create data frame of coordinates
heat_df = df_acc[['LATITUDE', 'LONGITUD']]
```

```python
# drop na values
heat_df = heat_df.dropna(axis=0, subset=['LATITUDE','LONGITUD'])
```

```python
# List comprehension to make out list of lists
heat_data = [[row['LATITUDE'],row['LONGITUD']] for index, row in heat_df.iterrows()]
```

```python
# Plot it on the map
HeatMap(heat_data).add_to(map_hooray)
```
